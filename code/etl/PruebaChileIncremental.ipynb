```python
def FuncionIDSismo(Dataset):
    for i in range(len(Dataset)):
        val1 = Dataset["Fecha_del_sismo"][i].strftime("%Y%m%d")
        val2 = Dataset["Hora_del_sismo"][i].strftime("%H")
        val3 = Dataset["Latitud"][i][:5].replace(".", "")
        val4 = Dataset["Longitud"][i][:5].replace(".", "")
        Dataset.loc[i,"ID_Sismo"] = val1+val2+val3+val4
    return Dataset
```
