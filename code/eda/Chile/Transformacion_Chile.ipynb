A continuación, la transformación del Dataset obtenido a partir de WebScrapping y las transformaciones para correlacionarlo con el primer Dataset.

```python
#Para importar nuestro segundo Dataset con sismos entre 2015 y 2023:
SismosChile2 = pd.read_csv("Datasets/SismosChile2015-2023.csv")
```

```python
#Para borrar las filas nulas:
SismosChile2 = SismosChile2.dropna(how='all')
```

En el Dataset obtenido quedaron columnas anidadas por la estructura de la web.

Las columnas Fecha local y Referencia se encuentran anidadas en una columna y Latitud y Longitud en otra.

Para desanidarlas separé las cadenas de caracteres en base a la cantidad de dígitos.

```python
#Para desanidar las columnas 1 y 3:
SismosChile2["Fecha local"] = SismosChile2["1"].str[:19]

SismosChile2["Referencia"] = SismosChile2["1"].str[19:]

SismosChile2["Latitud"] = SismosChile2["3"].str[:7]

SismosChile2["Longitud"] = SismosChile2["3"].str[8:]
```

```python
#Para borrar las columnas 1 y 3:
SismosChile2 = SismosChile2.drop(["1", "3"], axis=1)
```

A continuación renombramos las columnas y quitamos la nomenclatura km de la variable Profundidad.

```python
#Para renombrar las columnas 2, 4 y 5:
SismosChile2 = SismosChile2.rename(columns={"2":"Fecha UTC", "4":"Profundidad(Km)", "5":"Magnitud"})
```

```python
#Para quitar la nomenclatura km de la columna Profundidad:
SismosChile2["Profundidad(Km)"] = SismosChile2["Profundidad(Km)"].str.replace("km", "")
```

```python
SismosChile2
```

Por último, para poder unir ambos Dataset, respetamos el orden de columnas del primero, por lo que hacemos un index de columnas al segundo con los valores de columns del primero.

```python
#Para normalizar el índice de las dos columnas:
Index = SismosChile1.columns

SismosChile2 = SismosChile2.reindex(columns=Index)
```

Concatenamos ambos Dataset, reseteamos index de filas, normalizamos fechas transformando a Datetime y transformamos la variable Profundidad a float.

```python
#Para concatenar ambos Datasets:
SismosChile = pd.concat([SismosChile1, SismosChile2])
```

```python
SismosChile
```

```python
#Para resetear el índice de las filas:
SismosChile = SismosChile.reset_index(drop=True)
```

```python
#Para normalizar las columnas Datetime:
SismosChile["Fecha local"] = pd.to_datetime(SismosChile["Fecha local"])

SismosChile["Fecha UTC"] = pd.to_datetime(SismosChile["Fecha UTC"])
```

```python
#Para transformar Profundidad a float:
SismosChile["Profundidad(Km)"] = SismosChile["Profundidad(Km)"].astype(float)
```

```python
SismosChile = SismosChile.rename(columns={'Referencia': 'Lugar del Epicentro'})
```

```python
SismosChile
```

A continuación, el código para desanidar la columna Magnitud, creando una variable para cada tipo de magnitud y completandola con su valor correspondiente, utilizando la libreria Re.

```python
SismosChile["Ml"] = SismosChile["Magnitud"].apply(lambda x: float(re.findall(r'(\d+(?:\.\d+)?) Ml', x)[0]) if re.findall(r'(\d+(?:\.\d+)?) Ml', x) else None)

SismosChile["Mc"] = SismosChile["Magnitud"].apply(lambda x: float(re.findall(r'(\d+(?:\.\d+)?) Mc', x)[0]) if re.findall(r'(\d+(?:\.\d+)?) Mc', x) else None)

SismosChile["Mb"] = SismosChile["Magnitud"].apply(lambda x: float(re.findall(r'(\d+(?:\.\d+)?) Mb', x)[0]) if re.findall(r'(\d+(?:\.\d+)?) Mb', x) else None)

SismosChile["Ms"] = SismosChile["Magnitud"].apply(lambda x: float(re.findall(r'(\d+(?:\.\d+)?) Ms', x)[0]) if re.findall(r'(\d+(?:\.\d+)?) Ms', x) else None)

SismosChile["Mw"] = SismosChile["Magnitud"].apply(lambda x: float(re.findall(r'(\d+(?:\.\d+)?) Mw', x)[0]) if re.findall(r'(\d+(?:\.\d+)?) Mw', x) else None)
```

El siguiente código tiene la intención de obtener un promedio de las magnitudes

```python
SismosChile["MagnitudPromedio"] = 0 #Creamos la columna MagnitudPromedio con valores 0
lista = ("Ml", "Mc", "Mb", "Ms", "Mw") #Creamos una lista con los nombres de las magnitudes

for i in range(len(SismosChile)): #Iteramos todos los elementos del Dataframe
    sumatoria = 0 #Creamos las variables sumatoria y cantidad en 0
    cantidad = 0
    for elemento in lista: #Iteramos los nombres de las magnitudes almacenando valores no nulos
        if not np.isnan(SismosChile.loc[i,elemento]):
            sumatoria += (SismosChile.loc[i,elemento]) #Sumamos el valor de la magnitud a la sumatoria
            cantidad += 1 #Sumamos uno a cantidad para hacer la división
        else:
            continue
    try:
        SismosChile.loc[i, "MagnitudPromedio"] = round((sumatoria/cantidad), 1) #Intentamos sumatoria/cantidad
    except ZeroDivisionError:
        SismosChile.loc[i, "MagnitudPromedio"] = np.nan #A menos que el valor de cantidad sea 0, en ese caso cargamos valor nulo
```

Para obtener otra columna con cada magnitud y su valor correspondiente

```python
SismosChile["Magnitudes"] = None #Creamos la columna Magnitudes con valores nulos

for i in range(len(SismosChile)): #Iteramos todos los elementos del Dataframe
    magnitudes = [] #Creamos lista vacia de magnitudes
    for elemento in lista: #Iteramos los nombres de las magnitudes
        if not np.isnan(SismosChile.loc[i,elemento]):
            magnitudes.append(elemento + "(" + str(SismosChile.loc[i,elemento])+")") #Si el elemento no es nulo, hacemos un append a magnitudes Mg(valor)
        else:
            continue
    SismosChile.loc[i, "Magnitudes"] = ', '.join(magnitudes) #Unimos los elementos separados por coma
```

Para eliminar las columnas sobrantes

```python
columnas = ["Magnitud", "Ml", "Mc", "Mb", "Ms", "Mw"]

SismosChile = SismosChile.drop(columnas, axis=1)
```

Para reordenar las columnas

```python
columnas = ["Fecha local", "Fecha UTC", "Latitud", "Longitud", "Profundidad(Km)", "MagnitudPromedio", "Magnitudes", "Lugar del Epicentro"]

SismosChile = SismosChile.reindex(columns=columnas)
```
